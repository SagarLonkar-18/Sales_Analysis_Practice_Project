Exploratory Data Analysis

Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)

plt.show()
```

```python
# From above graphs we can see that most of the buyers are females 
```

AGE

```python
# Without Hue , Hue will create a division between men and women in our example
sns.countplot(data = df, x = 'Age Group')
plt.show()
```

```python
# With Hue

ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age,hue = 'Age Group')

plt.show()
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

State

```python
# total number of orders from top 10 states

# head shows top 10 states we can change the value here
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue = 'Orders')

plt.show()

# we can see the top 10 states from where the sales are more 
```

```python
# total amount/sales from top 10 states

# head shows top 10 states we can change the value here
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount',hue = 'Amount')

plt.show()

# we can see the top 10 states from where the sales are more 
```

```python
# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

Marital Status

```python
ax = sns.countplot(data = df, x = 'Married',hue = 'Married')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Married', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Married',y= 'Amount', hue='Gender')

plt.show()

# the values 1 , 2 ,3 ,4 are in form of 1 to the power 7 , 2 to the power 7 ... 10 to the power 7 amounts
```

```python
# From above graphs we can see that most of the buyers are married (women) 
```

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',hue='Occupation')

plt.show()
```

```python
# From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue='Product_Category')

plt.show()
```

```python
# From above graphs we can see that most of the sold products are from Clothing but amount wise food amount is more 
```

Product ID

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',hue='Orders')

plt.show()
```
